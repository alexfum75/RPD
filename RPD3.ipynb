### Diff (residual for  HH & Ind)

```python

fig = plt.figure(figsize=(15, 10))


ax1 = fig.add_subplot(3, 2, 1)
ax2 = fig.add_subplot(3, 2, 2)
ax3 = fig.add_subplot(3, 2, 3)
ax4 = fig.add_subplot(3, 2, 4)
ax5 = fig.add_subplot(3, 2, 5)

df_ind_diff = pd.DataFrame ((df_ind_ORI['Ind']  - df_ind_1['Ind']))
df_ind_diff ['min'] = df_ind_ORI['min'] 
x = df_ind_diff['min']
y = df_ind_diff['Ind']
ax1.plot(x,y, marker='.', label='diff (Ind - Ind1) ', color='red')
ax1.grid(True)
ax1.legend()

df_ind_diff = pd.DataFrame ((df_ind_ORI['Ind']  - df_ind_2['Ind']))
df_ind_diff ['min'] = df_ind_ORI['min'] 
x = df_ind_diff['min']
y = df_ind_diff['Ind']
ax2.plot(x,y, marker='.', label='diff (Ind - Ind2) ',color='green')
ax2.grid(True)
ax2.legend()

df_ind_diff = pd.DataFrame ((df_ind_ORI['Ind']  - df_ind_3['Ind']))
df_ind_diff ['min'] = df_ind_ORI['min'] 
x = df_ind_diff['min']
y = df_ind_diff['Ind']
ax3.plot(x,y, marker='.', label='diff (Ind - Ind3) ', color='blue')
ax3.grid(True)
ax3.legend()

df_ind_diff = pd.DataFrame ((df_ind_ORI['Ind']  - df_ind_4['Ind']))
df_ind_diff ['min'] = df_ind_ORI['min'] 
x = df_ind_diff['min']
y = df_ind_diff['Ind']
ax4.plot(x,y, marker='.', label='diff (Ind - Ind4) ',color='yellow')
ax4.grid(True)
ax4.legend()

df_ind_diff = pd.DataFrame ((df_ind_ORI['Ind']  - df_ind_5['Ind']))
df_ind_diff ['min'] = df_ind_ORI['min'] 
x = df_ind_diff['min']
y = df_ind_diff['Ind']
ax5.plot(x,y, marker='.', label='diff (Ind - Ind5) ',color='brown')
ax5.grid(True)
ax5.legend()
```
